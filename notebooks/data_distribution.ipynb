```python
import utils.load as load
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
O = load.load_dataset(data_path='./refactor/data/proc/', min_sample_thr=10,)

sns.set(font_scale=1)
sns.set_style('ticks',{'axes.grid' : False})
plt.figure(figsize=(20,5))
for i,ttype in enumerate(O['well_sampled_sorted_t_types']):
    count = np.sum(O['cluster']==ttype)
    plt.bar(i,count,facecolor='silver',edgecolor='None')
    plt.text(i,count+20,str(count),horizontalalignment='center',verticalalignment='center',rotation=90)
ax = plt.gca()

ax.set_xticks(np.arange(O['well_sampled_sorted_t_types'].size))
ax.set_xticklabels(O['well_sampled_sorted_t_types'].tolist(),rotation=90)
ax.set_xlim(-1,O['well_sampled_sorted_t_types'].size)
ax.set_xlabel('')
sns.despine()
plt.tight_layout()
np.savetxt('./docs/Src_Fig_ED_2.csv', np.array(O['well_sampled_sorted_t_types']), 
           fmt='%s', delimiter=',', 
           newline='\n', header='', 
           footer='', comments='# ', encoding=None)
plt.savefig('./docs/Fig_ED_2.jpg',dpi=300)
```
